```python
# Quiz 1) 카드 뭉치 최대 조합
# 기대결과 : 24 / 32 / 28

def max_product(left_cards, right_cards):
    # 여기에 코드를 작성하세요
    max_value = 0
    for left in left_cards:
        for right in right_cards:
            max_value = max(max_value, left*right)
    return max_value
# 테스트 코드
print(max_product([1, 6, 5], [4, 2, 3]))
print(max_product([1, -9, 3, 4], [2, 8, 3, 1]))
print(max_product([-1, -7, 3], [-4, 3, 6]))
```

```python
# Quiz 2) 가까운 매장 찾기

# 제곱근 사용을 위한 sqrt 함수
from math import sqrt

# 두 매장의 직선 거리를 계산해 주는 함수
def distance(store1, store2):
    return sqrt((store1[0] - store2[0]) ** 2 + (store1[1] - store2[1]) ** 2)

# 가장 가까운 두 매장을 찾아주는 함수
def closest_pair(coordinates):
    # 여기에 코드를 작성하세요
    min_dis = distance(coordinates[0],coordinates[1])
    
    for first in coordinates:
        temp_coor = coordinates
        temp_coor.remove(first)
        for second in temp_coor:
            if distance(first,second) < min_dis:
                min_dis = distance(first,second)
                item1 = first
                item2 = second

    return [item1, item2]

# 테스트 코드
test_coordinates = [(2, 3), (12, 30), (40, 50), (5, 1), (12, 10), (3, 4)]
print(closest_pair(test_coordinates))
```

1~100의 합
=> 1~50의 합 / 51~100의 합
=> 1~8의 합
=> 1~4의 합 / 5~8의 합
=> 1~2의 합 / 3~4의 합
=> 1~1의 합 / 2~2의 합 (Base case)

```python
# Quiz 4) 1부터 n까지의 합
def consecutive_sum(start, end):
    # 여기에 코드를 작성하세요
    if start == end:
        return start
    median = int((start + end)/2)
    return consecutive_sum(start,median) + consecutive_sum(median+1,end)
    
    
# 테스트 코드
print(consecutive_sum(1, 10))
print(consecutive_sum(1, 100))
print(consecutive_sum(1, 253))
print(consecutive_sum(1, 388)) 
```

- Conquer : 피봇 좌우 값을 정렬
- Combine : 자동으로 정렬되어 있음

```python
# Quiz 8) Partition 함수 구현하기

# 두 요소의 위치를 바꿔주는 helper function
def swap_elements(my_list, index1, index2):
    # a, b = b, a
    my_list[index1], my_list[index2] = my_list[index2], my_list[index1]
    return my_list

# 퀵 정렬에서 사용되는 partition 함수
def partition(my_list, start, end):
    # 여기에 코드를 작성하세요
    b = start
    i = 0

    for i in range(start,end):
        if my_list[i] <= my_list[end]:
            swap_elements(my_list, b, i)
            b += 1
            i += 1
        else:
            i += 1
    swap_elements(my_list, b, end)
    
    return b # pivot index

# 테스트 코드 1
list1 = [4, 3, 6, 2, 7, 1, 5]
pivot_index1 = partition(list1, 0, len(list1) - 1)
print(list1)
print(pivot_index1)

# 테스트 코드 2
list2 = [6, 1, 2, 6, 3, 5, 4]
pivot_index2 = partition(list2, 0, len(list2) - 1)
print(list2)
print(pivot_index2)
```

Dynamic Programming
- 최적 부분 구조(Optimal Substructure)
    . 피보나치 수 : fib(5) 해결을 위해서는 fib(4) 와 fib(3) 부분문제의 최적의 답을 이용해서 구할 수 있음 => 최적 부분 구조
    . 최단경로찾기 : 중간 경유지까지의 최단 거리 + 경유지에서 목적지까지 최단 거리 => 최적 부분 구조
- 중복되는 부분 문제(Overlapping Subproblems)
    . 피보나치 수: 
        fib(5)
        -> fib(4) + fib(3)
        -> (fib(3) + fib(2)) + fib(3)
        -> ((fib(2) + fib(1)) + fib(2)) + (fib(2) + fib(1))
    . 합병정렬 : 왼쪽/오른쪽 부분문제는 중복되지 않는 부분문제

- 최적 부분 구조가 있다?
    -> 기존문제를 부분문제로 나눠 풀 수 있다.
    -> 중복되는 부분 문제들이 있을 수 있다.
    -> 이를 해결하기 위해 Dynamic Programming : 한번 계산한 결과를 재활용

1. Memoization (하향식 접근)
    . 중복되는 계산은 한 번만 계산 후 메모
    . 한 번 계산한 내용을 cache 에 저장       


```python
# Quiz 11) 피보나치 수열 Memoization

def fib_memo(n, cache):
    # 여기에 코드를 작성하세요
    if n <= 1:
        return n
    
    if n in cache:
        return cache[n]
    
    cache[n] = fib_memo(n-1, cache) + fib_memo(n-2, cache)
    
    return cache[n]

def fib(n):
    # n번째 피보나치 수를 담는 사전
    fib_cache = {}
    return fib_memo(n, fib_cache)

# 테스트 코드
print(fib(10))
print(fib(50))
print(fib(100))
```

2. Tabulation (상향식 접근)
    . 중복되는 것부터 푸는 방법
    . fib(1) -> ... fib(5)
    . fib(1)=1, fib(2)=1, fib(3) = fib(1) + fib(2)...
    . n vs fib(n) 의 Table 을 채워나가는 방식으로 접근
    . Tabulation : Table 방식으로 정리
    . Memoization(재귀) vs Tabulation(반복문)

```python
# Quiz 12) 피보나치 수열 Tabulation

# 모범 답안
# def fib_tab(n):
#     # 이미 계산된 피보나치 수를 담는 리스트
#     fib_table = [0, 1, 1]

#     # n번째 피보나치 수까지 리스트를 하나씩 채워 나간다
#     for i in range(3, n + 1):
#         fib_table.append(fib_table[i - 1] + fib_table[i - 2])

#     # 피보나치 n번째 수를 리턴한다
#     return fib_table[n]

def fib_tab(n):
    # 여기에 코드를 작성하세요
    table = {}
    
    for i in range(1,n+1):
        if i <= 2:
            table[i] = 1
        else:
            table[i] = table[i-2] + table[i-1]
    return table[n]

# 테스트 코드
print(fib_tab(10))
print(fib_tab(56))
print(fib_tab(132))
```

Memoization vs Tabulation
- 공통점
    . 중복되는 부분 문제의 비효율을 해결
- 차이점
    . Memoization : 재귀함수 -> 필요한 계산만 수행, call stack overflow 위험
    . Tabulation : 반복문 -> n번 계산 필요, 중간에 필요없는 계산 포함

Tabulation 으로 피보나치 함수
- 기존 방식
    . 시간 복잡도 : O(n)
    . 공간 복잡도 : O(n)
- 공간 최적화 : current, previous 만 저장
    . 공간 복잡도 : O(1)

```python
# Quiz 13) 피보나치 수열 공간 최적화 (Tabulation)

# 모범 답안

# def fib_optimized(n):
#     current = 1
#     previous = 0

#     # 반복적으로 위 변수들을 업데이트한다. 
#     for i in range(1, n):
#         current, previous = current + previous, current

#     # n번재 피보나치 수를 리턴한다. 
#     return current


def fib_optimized(n):
    # 여기에 코드를 작성하세요
    current = 1
    previous = 0
    
    for i in range(1,n+1):
        if i <= 2:
            previous = 0
        result = current + previous
        previous = current
        current = result
    return result

# 테스트 코드
print(fib_optimized(3))
print(fib_optimized(16))
print(fib_optimized(53))
print(fib_optimized(213))
```

```python
# Quiz 15) 새꼼달꼼 장사 Tabulation

def max_profit(price_list, count):
    # 개수별로 가능한 최대 수익을 저장하는 리스트
    # 새꼼달꼼 0개면 0원
    table = [0]

    # 개수 1부터 count까지 계산하기 위해 반복문
    for i in range(1, count + 1):
        # count개를 팔 수 있는 조합들을 비교해서, 가능한 최대 수익을 찾는다
        
        if i < len(price_list):
            profit = price_list[i]
        else:
            profit = 0

        for j in range(1, i//2+1):
            profit = max(profit, table[i-j] + table[j])
        # 가능한 최대 수익을 profit_table에 넣는다
        table.append(profit)

    return table[count]


# 테스트 코드
print(max_profit([0, 100, 400, 800, 900, 1000], 5))
print(max_profit([0, 200, 600, 900, 1200, 2000], 5))
print(max_profit([0, 300, 600, 700, 1100, 1400], 8))
print(max_profit([0, 100, 200, 400, 600, 900, 1200, 1300, 1500, 1800], 9))

```

Greedy Algorithm : 당장 눈 앞에 보이는 최적의 선택을 하는 방식
- 장점 : 간단하고 빠르다
- 단점 : 최적의 답이 보장되지 않는다

Greedy Algorithm이 최적의 답을 보장해주는 경우
1. 최적 부분 구조를 가지는 경우
2. 탐욕적 선택 속성 : 각 단계에서의 탐욕스런 선택이 최종 답을 구하기 위한 최적의 선택

ex) 500/100/50/10원 최대한 적은 동전으로 돈 거슬러주기 : 1700원 거슬러주는 예시
1. 최적 부분 구조 : 500원 + 1200원 or 100원 + 1600원 or 50원 + 1650원 or 10원 + 1690원 부분문제
2. 탐욕적 선택 속성 : 매 순간 최대한 큰 동전을 선택 -> 큰 동전을 선택하는 것이 가장 좋은 선택임을 증명


```python
# Quiz 16) 최소 동전으로 거슬러 주기

def min_coin_count(value, coin_list):
    # 여기에 코드를 작성하세요

    sorted_coin = sorted(coin_list,reverse=True)
    cnt = 0
    for coin in sorted_coin:
        cnt = cnt + value//coin
        value = value%coin
    return cnt

# 테스트 코드
default_coin_list = [100, 500, 10, 50]
print(min_coin_count(1440, default_coin_list))
print(min_coin_count(1700, default_coin_list))
print(min_coin_count(23520, default_coin_list))
print(min_coin_count(32590, default_coin_list))
```

```python
# Quiz 17) 최대 곱 구하기
# 최적 부분 구조 : 1~n-1 까지의 곱의 최적값을 n번째와 조합하는 부분 문제로 나눌 수 있음
# 탐욕적 선택 속성 : 각 항목은 양수이므로 최대값을 선택하면, 곱의 최대값 보장됨

def max_product(card_lists):
    # 여기에 코드를 작성하세요
    product = 1
    for card in card_lists:
        product = product * max(card)
    
    return product


# 테스트 코드
test_cards1 = [[1, 6, 5], [4, 2, 3]]
print(max_product(test_cards1))

test_cards2 = [[9, 7, 8], [9, 2, 3], [9, 8, 1], [2, 8, 3], [1, 3, 6], [7, 7, 4]]
print(max_product(test_cards2))

test_cards3 = [[1, 2, 3], [4, 6, 1], [8, 2, 4], [3, 2, 5], [5, 2, 3], [3, 2, 1]]
print(max_product(test_cards3))

test_cards4 = [[5, 5, 5], [4, 3, 5], [1, 1, 1], [9, 8, 3], [2, 8, 4], [5, 7, 4]]
print(max_product(test_cards4))
```

```python
# Quiz 18) 지각 벌금 적게 내기
# 최적 부분 구조 : 1~5 경우에 대해 최소값을 찾고, 그 다음 단계의 부분문제
# 탐욕적 선택 속성 : 매 단계 최소값을 선택하는 것이 최선의 답을 보장

def min_fee(pages_to_print):
    # 여기에 코드를 작성하세요
    person = len(pages_to_print)
    sorted_pages = sorted(pages_to_print)
    fee = 0
    for page in sorted_pages:
        fee = fee + page * person
        person = person - 1
    
    return fee

# 테스트 코드
print(min_fee([6, 11, 4, 1]))
print(min_fee([3, 2, 1]))
print(min_fee([3, 1, 4, 3, 2]))
print(min_fee([8, 4, 2, 3, 9, 23, 6, 8]))

```
